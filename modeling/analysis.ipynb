### 데이터 불러오기

```python
import shap
import pickle
import pandas as pd
```

```python
with open('model1.pickle','rb') as file:
    model = pickle.load(file)
```

```python
data = pd.read_csv('dataset/modeldata.csv')
target = data['price_gold']
feature = data.drop(['price_gold','date','call','index'],axis=1)
```

### Shap

```python
explainer = shap.Explainer(model)
shap_val = explainer(feature)
shap.plots.beeswarm(shap_val)
```

```python
shap.plots.waterfall(shap_val[50])
```
