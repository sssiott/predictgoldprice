```python
import shap
import pickle
import pandas as pd
```

```python
with open('model.pickle','rb') as file:
    model = pickle.load(file)
```

```python
data = pd.read_csv('dataset.csv')
target = data['price_gold']
feature = data.drop('price_gold',axis=1)
```

```python
explainer = shap.Explainer(model)
```

```python
shap_val = explainer(feature)
```

```python
shap.plots.beeswarm(shap_val)
```

```python
shap.plots.waterfall(shap_val[50])
```
